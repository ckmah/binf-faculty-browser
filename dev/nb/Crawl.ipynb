# Crawl and Compile Faculty Listings

```python
import scrapy
from scrapy.crawler import CrawlerProcess

import os
import sys
sys.path.append('../crawler/crawler/spiders/')

from faculty_spider import FacultySpider
```

```python
DATA_DIR = '../data/'
```

## Scrape for faculty information

```python
%%bash
rm ../data/faculty.jl
```

```python
# Initialize crawler
process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)',
})

# Settings
file_format = 'jl'
file_path = '{}faculty.{}'.format(DATA_DIR, file_format)
process.settings.set('FEED_FORMAT', file_format)
process.settings.set('FEED_URI', file_path)

process.crawl(FacultySpider)
process.start()
```
